```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"superstore sales dataset.csv")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#calculating number of null values 
null_count = df["Postal Code"].isnull().sum()
null_count
```

```python
#replacing null values with 0
df["Postal Code"].fillna(0,inplace=True)

#converting float to integer datatype
df["Postal Code"] = df["Postal Code"].astype(int)
df.info()
```

```python
df.describe()
```

```python
#count of duplicate values
df.duplicated(keep = False).sum()
```

```python
types_of_customers = df["Segment"].unique()
print(types_of_customers)
```

```python
number_of_customers = df["Segment"].value_counts().reset_index()

number_of_customers = number_of_customers.rename(columns={'Segment':'customer_type','count':'customer_count'})
print(number_of_customers)
```

```python
#plotting a pie chart
plt.pie(number_of_customers['customer_count'],labels=number_of_customers['customer_type'],autopct='%1.2f%%')

plt.title("Distribution of customers")
plt.show()
```

```python

# Grouping by 'Segment' and summing 'Sales' within each segment
sales_per_segment = df.groupby('Segment')['Sales'].sum().reset_index()
sales_per_segment = sales_per_segment.rename(columns={'Segment':'customer_type','Sales':'total_sales'})
# Displaying sales per customer segment
print(sales_per_segment)
```

```python
#plotting a pie chart
plt.pie(sales_per_segment['total_sales'],labels=sales_per_segment['customer_type'],autopct='%1.2f%%')

plt.title("Distribution of sales")
plt.show()
```

```python
plt.bar(sales_per_segment['customer_type'],sales_per_segment['total_sales'])
plt.title('sales per customer category')
plt.xlabel('customer_type')
plt.ylabel('total_sales')
plt.show()
```

# Group data according to Customer ID,Customer Name,Segment

```python
customer_order_frequency = df.groupby(['Customer ID','Customer Name','Segment'])['Order ID'].count().reset_index()

#rename order id column
customer_order_frequency = customer_order_frequency.rename(columns={'Order ID':'Total_Orders'})

#identify repeated customer
repeated_customers = customer_order_frequency[customer_order_frequency['Total_Orders']>=1]

#sort customers in descending order
sorted_repeated_customer = repeated_customers.sort_values(by='Total_Orders',ascending = False)

print(sorted_repeated_customer.head(10).reset_index(drop = True))
```

# Ranking customers According to Sales

```python

# Grouping by 'Customer ID' and summing 'Sales' for each customer
customer_sales = df.groupby(['Customer ID','Customer Name','Segment'])['Sales'].sum().reset_index()

customer_sales['Sales_Rank'] = customer_sales['Sales'].rank(ascending=False)

# Displaying the ranked customers according to sales
print(customer_sales.sort_values(by='Sales', ascending=False).head(10))
```

# Mode of Shipping

# Sales

```python
# Convert the "Order Date" column to datetime format

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Group the data by years and calculate the total sales amount for each year
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum().reset_index()


yearly_sales = yearly_sales.rename(columns={'Order Date': 'Year', 'Sales':'Total_Sales'})

 
# Print the total sales for each year
print(yearly_sales)
```

```python
# Ploting a bar graph

plt.bar(yearly_sales['Year'], yearly_sales['Total_Sales'])

# Labels
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total_Sales')
plt.xticks(rotation=45)

plt.show()
```

```python
# Create a line graph for total sales by year
plt.plot(yearly_sales['Year'], yearly_sales['Total_Sales'], marker='o', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Total_Sales')
plt.title('Total Sales by Year')

# Display the plot
plt.tight_layout()

plt.show()
```

```python
# Filter the data for the year 2018
year_sales = df[df['Order Date'].dt.year == 2018]

# Calculate the quarterly sales for 2018
quarterly_sales = year_sales.resample('Q', on='Order Date')['Sales'].sum().reset_index()


quarterly_sales = quarterly_sales.rename(columns={'Order Date': 'Quarter', 'Sales':'Total_Sales'})


print("Quarterly Sales for 2018:")
print(quarterly_sales)
```

```python
# Create a line graph for total sales by year 2018
plt.plot(quarterly_sales['Quarter'], quarterly_sales['Total_Sales'], marker='o', linestyle='--')
plt.xlabel('Quarter')
plt.ylabel('Total_Sales')
plt.title('Total Sales by Year 2018')

# Display the plot
plt.tight_layout()
plt.xticks(rotation=75)

plt.show()
```

```python

# Calculate the monthly sales for 2018
monthly_sales = year_sales.resample('M', on='Order Date')['Sales'].sum().reset_index()

# Renaming the columns

monthly_sales = monthly_sales.rename(columns={'Order Date':'Month', 'Sales':'Total_Montly_Sales'})

# Print the monthly and quarterly sales for 2018
print("Monthly Sales for 2018:")
print(monthly_sales)
```

```python
# Create a line graph for total sales by year 2018
plt.plot(monthly_sales['Month'], monthly_sales['Total_Montly_Sales'], marker='o', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Total_Montly_Sales')
plt.title('Total Sales by Year 2018')

# Display the plot
plt.tight_layout()


plt.show()
```

```python
# Group the data by months and calculate the total sales amount for each month
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

# Plot the sales trends for months
plt.figure(figsize=(12, 26))

# Monthly Sales Trend
plt.subplot(3, 1, 1)
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales Amount')

# Adjust layout and display the plots
# plt.tight_layout()
plt.show()
```

```python
# Group the data by years and calculate the total sales amount for each year
yearly_sales = df.groupby(df['Order Date'].dt.to_period('Y'))['Sales'].sum()

# Plot the sales trends for quarters
plt.figure(figsize=(12, 26))

# Yearly Sales Trend
plt.subplot(3, 1, 3)
yearly_sales.plot(kind='line', marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales Amount')

# Adjust layout and display the plots

plt.show()
```
